# 01 — Data Ingestion
Pull the 2024 F1 Constructor Standings via FastF1's Ergast API.

```python
from fastf1.ergast import Ergast

ergast = Ergast()
response = ergast.get_constructor_standings(season=2024)

# Final-round standings (last entry = end-of-season)
standings = response.content[-1]
standings
```

```python
display_cols = ['position', 'constructorName', 'constructorNationality', 'points', 'wins']
standings[display_cols].reset_index(drop=True)
```
